**Data Summary**

ADVENTUREWORKS DATASET

The AdventureWorks dataset is a sample dataset created by Microsoft. It represents a fictional retail company that sells bicycles and related products. This dataset represents a complete sales ecosystem of a retail business.
Although the data is not real, it follows a structure similar to actual company databases, including sales transactions, customer details, product hierarchy, and regional information.

What this dataset is about?

This dataset focuses on sales operations of the company. It allows analysis of:

- Sales performance over time
- Product and category performance
- Customer purchasing behavior
- Regional sales distribution
- Product returns and their impact

From an initial look, the dataset contains multiple related tables, indicating a relational data structure where different aspects of the business are stored separately and connected through keys.

Data Structure Overview

The dataset is organized into:

- Fact Table (Main Data): Sales transactions
- Dimension Tables (Supporting Data): Customer, Product, Calendar, Territory, etc.

The sales data is split across multiple files (2020â2022), which need to be combined into a single table for analysis.

Tables Overview

- Sales Data (2020, 2021, 2022)
  - Transaction-level data including product, customer, date, quantity, and sales amount. These tables form the core dataset.

- Returns Data
  - Contains records of returned products, useful for analyzing return rates and product issues.

- Calendar Lookup
  - Provides date-related details such as year, month, and quarter for time-based analysis.

- Customer Lookup
  - Contains customer-related information, including location and demographics.

- Product Lookup
  - Includes product-level details.

- Product Subcategories Lookup
  - Organizes products into subcategories.

- Product Categories Lookup
  - Defines broader product categories.

- Territory Lookup
  - Contains geographic information such as region and country.

   Initial Observations

- The dataset is multi-table and relational, requiring joins for full analysis
- Sales data is time-based and split across years, requiring consolidation
- Product data follows a hierarchical structure (Category â Subcategory â Product)
- Presence of returns data allows net sales and return analysis
- Geographic data enables regional performance analysis

**Decription of each columns of the Dataset**

 Sales Tables (Sales_2020, Sales_2021, Sales_2022): 

- Order Date: The date when the customer placed the order, used for time-based sales analysis.
- Stock Date: The date when the ordered product was dispatched or became available from inventory.
- Order Number: A unique identifier assigned to each sales order transaction.
- Product Key: A unique identifier linking each record to a specific product in the product table.
- Customer Key: A unique identifier representing the customer who made the purchase.
- Territory Key: A unique identifier that links the sale to a specific sales region or territory.
- Order Line Item: The line number within an order indicating multiple products in a single order.
- Order Quantity: The total number of units of the product ordered in that transaction line.

Returns Table:

- Return Date: The date on which the product was returned by the customer.
- Territory Key: Unique identifier for the sales region where the return occurred; links to the Territory table.
- Product Key: Unique identifier of the returned product; links to the Product table.
- Return Quantity: The number of units of the product that were returned.

 Customer Table:

- Customer Key: Unique identifier assigned to each customer in the dataset.
- Prefix: Title used before the customerâs name.
- First Name: Customerâs given name.
- Last Name: Customerâs family or surname.
- Birth Date: Customerâs date of birth, used to calculate age.
- Marital Status: Indicates whether the customer is married or single.
- Gender: Gender of the customer (e.g., Male, Female).
- Email Address: Customerâs email contact information.
- Annual Income: Estimated yearly income of the customer.
- Total Children: Total number of children the customer has.
- Educational Level: Highest level of education attained by the customer.
- Occupation: Customerâs job or profession category.
- Homeowner: Indicates whether the customer owns a home (Yes/No).

Product Table:

- Product Key: A unique identifier assigned to each product. It is used to link product information across different tables.
- Product Subcategory Key: Identifies the subcategory to which the product belongs. It helps group products into more specific categories.
- Product SKU: The Stock Keeping Unit, a unique code used for tracking and managing inventory.
- Product Name: The official name of the product as displayed to customers.
- Model Name: Specifies the model or version of the product, often used to differentiate similar products.
- Product Description: A brief textual description providing details about the productâs features and characteristics.
- Product Color: Indicates the color of the product.
- Product Size: Represents the size or dimensions of the product, if applicable.
- Product Style: Describes the design or style category of the product.
- Product Cost: The cost incurred to manufacture or procure the product.
- Product Price: The selling price of the product offered to customers.

Product Category Table:

- ProductCategoryKey: Unique ID for each category.
- CategoryName: Name of the product category

Product Sub-category Table:

- ProductSubcategoryKey â Unique ID for each subcategory.
- SubcategoryName â Name of the subcategory.
- ProductCategoryKey â Links subcategory to its category.

Region/Territory Table:

- SalesTerritoryKey: Unique identifier for each sales territory.
- Region: The specific region name (e.g., Northeast, Southwest).
- Country: The country where the territory is located.
- Continent: The continent of the sales territory.

Calendar Table:

- Date: Stores the calendar date in YYYY-MM-DD format. It represents each individual day and is used to connect and analyze time-based data across the dataset.

**DATA CLEANING**

```python
# Making copies of the dataframes to work with

calendar_copy = calendar.copy()
sales_2020_copy = sales_2020.copy()
sales_2021_copy = sales_2021.copy()
sales_2022_copy = sales_2022.copy()
product_copy = product.copy()
customer_copy = customer.copy()
product_category_copy = product_category.copy()
product_subcategory_copy = product_subcategory.copy()
returns_copy = returns.copy()
region_copy = region.copy()
```

```python
# Changing the data type of the relevant columns of sales_2020_copy table.

sales_2020_copy["OrderDate"] = pd.to_datetime(sales_2020_copy["OrderDate"], errors="coerce")
sales_2020_copy["StockDate"] = pd.to_datetime(sales_2020_copy["StockDate"], errors="coerce")
sales_2020_copy = sales_2020_copy.astype({"ProductKey": "string", "CustomerKey": "string", "TerritoryKey": "string"})
```

```python
# Changing the data type of the relevant columns of sales_2021_copy table.

sales_2021_copy["OrderDate"] = pd.to_datetime(sales_2021_copy["OrderDate"], errors="coerce")
sales_2021_copy["StockDate"] = pd.to_datetime(sales_2021_copy["StockDate"], errors="coerce")
sales_2021_copy = sales_2021_copy.astype({"ProductKey": "string", "CustomerKey": "string", "TerritoryKey": "string"})
```

```python
# Changing the data type of the relevant columns of sales_2022_copy table.

sales_2022_copy["OrderDate"] = pd.to_datetime(sales_2022_copy["OrderDate"], errors="coerce")
sales_2022_copy["StockDate"] = pd.to_datetime(sales_2022_copy["StockDate"], errors="coerce")
sales_2022_copy = sales_2022_copy.astype({"ProductKey": "string", "CustomerKey": "string", "TerritoryKey": "string"})
```

```python
# Changing the data type of the relevant columns of returns_copy table.

returns_copy["ReturnDate"] = pd.to_datetime(returns_copy["ReturnDate"], errors="coerce")
returns_copy = returns_copy.astype({"ProductKey": "string", "TerritoryKey": "string"})
```

```python
# Changing the data type of the relevant columns of customer_copy table.

customer_copy["TotalChildren"] = customer_copy["TotalChildren"].astype("Int64")
customer_copy["BirthDate"] = pd.to_datetime(customer_copy["BirthDate"], errors="coerce")
```

```python
# Handling the missing values in the customer_copy table.

customer_copy = customer_copy.dropna(subset=["FirstName", "LastName"])
customer_copy = customer_copy.fillna({"Prefix": "Unknown", "Gender": "Unknown"})
```

```python
# Standardized categorical values

customer_copy['MaritalStatus'] = customer_copy['MaritalStatus'].replace({'M': 'Married','S': 'Single'})
customer_copy['Gender'] = customer_copy['Gender'].replace({'M': 'Male','F': 'Female'})
customer_copy['HomeOwner'] = customer_copy['HomeOwner'].replace({'Y': 'Yes','N': 'No'})
customer_copy["Occupation"] = customer_copy["Occupation"].str.strip().str.title()
customer_copy["EmailAddress"] = customer_copy["EmailAddress"].str.strip()
customer_copy["EducationLevel"] = customer_copy["EducationLevel"].str.strip().str.title()
```

```python
# Feature engineering for the customer_copy table.
customer_copy["CustomerName"] = customer_copy["FirstName"] + " " + customer_copy["LastName"]
customer_copy["CustomerName"] = customer_copy["CustomerName"].str.title()
customer_copy["CustomerName"] = customer_copy["CustomerName"].str.strip()
```

```python
customer_copy.insert(1, 'CustomerName', customer_copy.pop('CustomerName'))
customer_copy.drop(columns=['Prefix', 'FirstName', 'LastName'], inplace=True)
```

```python
# Changing the data type of the relevant columns of product_copy table.

product_copy = product_copy.astype({"ProductKey": "string", "ProductSubcategoryKey": "string"})
```

```python
# Handling the missing values in the product_copy table.

product_copy["ProductColor"] = product_copy["ProductColor"].fillna("Unknown")
product_copy["ProductDescription"] = product_copy["ProductDescription"].replace({"0": "No description available"})
product_copy["ProductSize"] = product_copy["ProductSize"].replace({"0": "Unknown"})
product_copy["ProductStyle"] = product_copy["ProductStyle"].replace({"0": "Unknown"})
```

```python
# Standardizing the relevant columns in the product_copy table.

product_copy["ProductColor"] = product_copy["ProductColor"].str.strip().str.title()
product_copy["ProductDescription"] = product_copy["ProductDescription"].str.strip()
product_copy["ProductName"] = product_copy["ProductName"].str.strip().str.title()
product_copy["ModelName"] = product_copy["ModelName"].str.strip().str.title()

product_copy["ProductStyle"] = product_copy["ProductStyle"].replace({"U": "Universal", "W": "Women's", "M": "Men's"})
```

```python
# Changing the data type of the relevant columns of product_category_copy table.

product_category_copy["ProductCategoryKey"] = product_category_copy["ProductCategoryKey"].astype("string")
```

```python
# Changing the data type of the relevant columns of product_Subcategory_copy table.

product_subcategory_copy["ProductCategoryKey"] = product_subcategory_copy["ProductCategoryKey"].astype("string")
product_subcategory_copy["ProductSubcategoryKey"] = product_subcategory_copy["ProductSubcategoryKey"].astype("string")
```

```python
# Changing the data type and column names of the relevant columns of region_copy table.

region_copy["SalesTerritoryKey"] = region_copy["SalesTerritoryKey"].astype("string")

region_copy = region_copy.rename(columns={"SalesTerritoryKey": "TerritoryKey"})
```

```python
# Changing the data type of the relevant columns of calendar_copy table.

calendar_copy["Date"] = pd.to_datetime(calendar_copy["Date"], errors="coerce")
```

```python
# creating a full calendar table with the relevant columns from the calendar_copy table.

calendar_copy["Year"] = calendar_copy["Date"].dt.year
calendar_copy["MonthName"] = calendar_copy["Date"].dt.month_name()
calendar_copy["Month"] = calendar_copy["Date"].dt.month
calendar_copy["DayName"] = calendar_copy["Date"].dt.day_name()
calendar_copy["Day"] = calendar_copy["Date"].dt.day
calendar_copy["Quarter"] = 'Q' + calendar_copy["Date"].dt.quarter.astype(str)
```
